```python
from wordcloud_fa import WordCloudFa
import numpy as np
import pandas as pd
from PIL import Image
from matplotlib.image import imread
import re
import matplotlib.pyplot as plt



with open('news_data.txt', 'r' , encoding='utf-8') as file:
  text = file.read()
mask_image = np.array(Image.open("twitter.png"))
Word_cloud = WordCloudFa(mask=mask_image, 
                       background_color="#f2f2f2",
                        max_font_size=100,
                        min_font_size=15,  
                        scale=10,
                        collocations=False,
                        no_reshape=False, 
                        remove_unhandled_utf_characters=True,  
                        persian_normalize=False, 
                        include_numbers=False )
Word_cloud.add_stop_words_from_file("stop_word.txt"),
frequencies = Word_cloud.process_text(text)
cloud = Word_cloud.generate_from_frequencies(frequencies)
print(len(text))

plt.imshow(cloud)
plt.axis("off")
plt.show()
```
