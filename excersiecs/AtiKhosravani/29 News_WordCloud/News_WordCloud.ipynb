**generate word cloud for each news**

تمام اخبار را با جدا کننده "-------------------" جدا میکنیم و کد برای هر یک اجرا شده و ابر کلمه هرکدام را به صورت جداگانه نمایش میدهد. به دلیل طولانی بودن کلیه اخبار از 19 خبر به عنوان نمونه استفاده شده و برای باقی نیز همین کد تعمیم داده میشود.

```python
import wordcloud_fa
import matplotlib.pyplot as plt
import PIL

word_cloud = WordCloudFa(height= 1080 , width=1920, background_color='white', colormap='Set2')
with open('news.txt' , 'r') as file:
  all_news_content = file.read()

news_array = all_news_content.split("-------------------");
i = 1;
for single_news in news_array:
  cloud = word_cloud.generate(single_news)
  plt.imshow(word_cloud)
  plt.axis('off')
  plt.title(i);
  plt.show()
  i+=1;
```
