```python
from wordcloud_fa import WordCloudFa
from string import *
import numpy as np
import pandas as pd
from PIL import Image
from matplotlib.image import imread
import re
import matplotlib.pyplot as plt
```

```python
foring_alfabet = ['C', 'V', 'B', 'N', 'M', 'Q', 'W', 'E', 'R', 'T', 'Y', 'U', 'I', 'O', 'P', 'A', 'S', 'D', 'F', 'G',
                  'H', 'J', 'K', 'L', 'Z', 'X', 'q', 'w', 'e', 'r', 't', 'y', 'u', 'i', 'o', 'p', 'a', 's', 'd', 'f',
                  'g', 'h', 'j', 'k', 'l', 'z', 'x', 'c', 'v', 'b', 'n', 'm']
stop = []
Xt: str = ''
```

```python
with open("stop-words.txt", encoding="utf8") as q1:
    for i6 in q1:
        stop.append(i6)
for u in range(0, len(stop)):
    p = stop[u]
    w = p.replace('\u200c', ' ')
    s = w.replace('\u200f', ' ')
    stop[u] = s.replace('\n', '')

with open("news1.txt", "r+", encoding="utf8") as K:
    for i in K:
        X = str.split(i)
        for u1 in range(0, len(X)):
            p1 = X[u1]
            w1 = p1.replace('\u200c', '')
            s1 = w1.replace('\u200f', '')
            X[u1] = s1.replace('\n', '')
        for i1 in range(0, len(X)):
            q2 = X[i1]
            M = ''.join([i for i in q2 if not i.isdigit()])
            X[i1] = M

        for i2 in foring_alfabet:
            for i3 in range(0, len(X)):
                q3 = X[i3]
                if i2 not in q3:
                    continue
                else:
                    X[i3] = q3.replace(i2, '')

        for i4 in punctuation:
            for i5 in range(0, len(X)):
                q = X[i5]
                if i4 not in q:
                    continue
                else:
                    X[i5] = q.replace(i4, '')

        for i7 in stop:
            if i7 not in X:
                continue
            else:
                X.remove(i7)
        for i10 in range(0, len(X)):
            Xt = X[i10] + ' ' + Xt
            Xtt = Xt + '\n'
        K2 = open("news1nuw.txt", "r+", encoding="utf8")
        K2.write(Xtt)
        K2.close()
```

```python
with open('news1nuw.txt', 'r' , encoding='utf-8') as file:
    text = file.read()
mask_image = np.array(Image.open("Iran_map_vector.svg.png"))
Word_cloud = WordCloudFa(mask=mask_image, 
                       background_color="#f2f2f2",
                        max_font_size=40,
                        min_font_size=15,  
                        scale=1,
                        collocations=False,
                        no_reshape=False, 
                        remove_unhandled_utf_characters=True,  
                        persian_normalize=False, 
                        include_numbers=False )
Word_cloud.add_stop_words_from_file("stop-words.txt"),
frequencies = Word_cloud.process_text(text)
cloud = Word_cloud.generate_from_frequencies(frequencies)
print("کل اخبار")
plt.imshow(cloud)
plt.show()
```
