```python
def xor(a, b):
  if a == False: 
      if b==False:
        return False
  if a == False: 
      if b==True:
        return True

  if a == True: 
      if b==True:
        return False
  if a == True: 
      if b==False:
        return True     

print("A", "B", "XOR")        
print("0", "0",xor(0,0))
print("0", "1",xor(0,1))
print("1", "0",xor(1,0))
print("1", "1",xor(1,1))
```

```python
import pandas
from sklearn import tree
import pydotplus
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
df = pandas.read_csv("/content/XOR")
print(df)
features = ['A', 'B']
X = df[features]
y = df['XOR']
print(df)
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)
data = tree.export_graphviz(dtree, out_file=None, feature_names=features)
graph = pydotplus.graph_from_dot_data(data)
graph.write_png('XOR.png')

img=pltimg.imread('XOR.png')
imgplot = plt.imshow(img)
plt.show()
```
