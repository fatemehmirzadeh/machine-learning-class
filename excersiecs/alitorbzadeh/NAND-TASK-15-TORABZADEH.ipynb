```python
def nand(a, b):
  if a == False:
      return True
  if a == True: 
      if b==False:
        return True
  if a == True: 
      if b==True:
        return False

print("A", "B", "NAND")        
print("0", "0", nand(0,0))
print("0", "1", nand(0,1))
print("1", "0", nand(1,0))
print("1", "1", nand(1,1))
```

```python
import pandas
from sklearn import tree
import pydotplus
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
df = pandas.read_csv("/content/NAND")
print(df)
features = ['A', 'B']
X = df[features]
y = df['NAND']
print(df)
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)
data = tree.export_graphviz(dtree, out_file=None, feature_names=features)
graph = pydotplus.graph_from_dot_data(data)
graph.write_png('NAND.png')

img=pltimg.imread('NAND.png')
imgplot = plt.imshow(img)
plt.show()
```
