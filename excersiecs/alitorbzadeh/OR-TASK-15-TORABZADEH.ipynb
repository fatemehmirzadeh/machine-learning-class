```python
def ORR(a, b):
  if a == False:
    if b == False:
      return False
    elif b == True: 
      return True
  elif a == True: 
      return True
print("A", "B", "OR")        
print("0", "0",ORR(0,0))
print("0", "1",ORR(0,1))
print("1", "0",ORR(1,0))
print("1", "1",ORR(1,1))
```

```python
import pandas
from sklearn import tree
import pydotplus
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
df = pandas.read_csv("/content/OR")
print(df)
features = ['A', 'B']
X = df[features]
y = df['OR']
print(df)
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)
data = tree.export_graphviz(dtree, out_file=None, feature_names=features)
graph = pydotplus.graph_from_dot_data(data)
graph.write_png('OR.png')

img=pltimg.imread('OR.png')
imgplot = plt.imshow(img)
plt.show()
```
